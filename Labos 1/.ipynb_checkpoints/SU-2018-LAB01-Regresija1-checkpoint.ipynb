### 1. Jednostavna regresija

Zadan je skup primjera $\mathcal{D}=\{(x^{(i)},y^{(i)})\}_{i=1}^4 = \{(0,4),(1,1),(2,2),(4,5)\}$. Primjere predstavite matrixom $\mathbf{X}$ dimenzija $N\times n$ (u ovom slučaju $4\times 1$) i vektorom oznaka $\textbf{y}$, dimenzija $N\times 1$ (u ovom slučaju $4\times 1$), na sljedeći način:


```python
X = np.array([[0],[1],[2],[4]])
y = np.array([4,1,2,5])
```

#### (a)

Proučite funkciju [`PolynomialFeatures`](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html) iz biblioteke `sklearn` i upotrijebite je za generiranje matrice dizajna $\mathbf{\Phi}$ koja ne koristi preslikavanje u prostor više dimenzije (samo će svakom primjeru biti dodane *dummy* jedinice; $m=n+1$).


```python
#generira matricu značajki koja se sastoji od svih kombinacija danih značajki do specificiranog stupnja = matrica dizajna

from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(1)
psi=poly.fit_transform(X)
```

#### (b)

Upoznajte se s modulom [`linalg`](http://docs.scipy.org/doc/numpy/reference/routines.linalg.html). Izračunajte težine $\mathbf{w}$ modela linearne regresije kao $\mathbf{w}=(\mathbf{\Phi}^\intercal\mathbf{\Phi})^{-1}\mathbf{\Phi}^\intercal\mathbf{y}$. Zatim se uvjerite da isti rezultat možete dobiti izračunom pseudoinverza $\mathbf{\Phi}^+$ matrice dizajna, tj. $\mathbf{w}=\mathbf{\Phi}^+\mathbf{y}$, korištenjem funkcije [`pinv`](http://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.pinv.html).

```python
from numpy import linalg
```

```python
#prvi nacin
temp = np.matmul(psi.transpose(), psi)
temp = linalg.inv(temp)
temp = np.matmul(temp, psi.transpose())
w1 = np.matmul(temp,y )

#drugi nacin
#pinv direktno izracunava pseudoinverz
w2=np.matmul(linalg.pinv(psi), y)

print(w1,w2)
```

Radi jasnoće, u nastavku je vektor $\mathbf{x}$ s dodanom *dummy* jedinicom $x_0=1$ označen kao $\tilde{\mathbf{x}}$.

#### (c)

Prikažite primjere iz $\mathcal{D}$ i funkciju $h(\tilde{\mathbf{x}})=\mathbf{w}^\intercal\tilde{\mathbf{x}}$. Izračunajte pogrešku učenja prema izrazu $E(h|\mathcal{D})=\frac{1}{2}\sum_{i=1}^N(\tilde{\mathbf{y}}^{(i)} - h(\tilde{\mathbf{x}}))^2$. Možete koristiti funkciju srednje kvadratne pogreške [`mean_squared_error`]( http://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html#sklearn.metrics.mean_squared_error) iz modula [`sklearn.metrics`](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics).

**Q:** Gore definirana funkcija pogreške $E(h|\mathcal{D})$ i funkcija srednje kvadratne pogreške nisu posve identične. U čemu je razlika? Koja je "realnija"?

```python
x_tilda = psi #nadodan dummy
w = np.reshape(w1,(-1,1)) #vektor u matricu sa jednom dimenzijom 1 kako bi se izvelo mnozenje s matricom
h = np.matmul(x_tilda, w)  
h = h.flatten()
E = 0.5*np.sum((y-h)**2)
print(E)


```

### 2. Polinomijalna regresija i utjecaj šuma

#### (a)

Razmotrimo sada regresiju na većem broju primjera. Koristite funkciju `make_labels(X, f, noise=0)` koja uzima matricu neoznačenih primjera $\mathbf{X}_{N\times n}$ te generira vektor njihovih oznaka $\mathbf{y}_{N\times 1}$. Oznake se generiraju kao $y^{(i)} = f(x^{(i)})+\mathcal{N}(0,\sigma^2)$, gdje je $f:\mathbb{R}^n\to\mathbb{R}$ stvarna funkcija koja je generirala podatke (koja nam je u stvarnosti nepoznata), a $\sigma$ je standardna devijacija Gaussovog šuma, definirana parametrom `noise`. Za generiranje šuma koristi se funkcija [`numpy.random.normal`](http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html). 

Generirajte skup za učenje od $N=50$ primjera uniformno distribuiranih u intervalu $[-5,5]$ pomoću funkcije $f(x) = 5 + x -2 x^2 -5 x^3$ uz šum  $\sigma=200$:

```python
from numpy.random import normal
def make_labels(X, f, noise=0) :
    return map(lambda x : f(x) + (normal(0,noise) if noise>0 else 0), X)
```

```python
def make_instances(x1, x2, N) :
    return np.array([np.array([x]) for x in np.linspace(x1, x2, N)])
#def make_instances(x1, x2, N):
#    instances = np.array([])
#    for x in np.linspace(x1, x2, N):
#        instances.append(x)
#    return instances
```

```python
def f(x):
    return 5+x-2*x**2-5*x**3
X = make_instances(-5, 5, 50)
#y= make_labels(X, f, 200)
#print(y)
y = list(make_labels(X, f, 200))  #ovaj map se ponaas kao neki generator ne vrati sve odjednom nego kako je pozivbas vraca iduci elemnt, pa ga se ocvako natira da vrati sve
y = np.array(y)
print(y)
```

Prikažite taj skup funkcijom [`scatter`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter).

```python
plt.scatter(X,y)
```

#### (b)

Trenirajte model polinomijalne regresije stupnja $d=3$. Na istom grafikonu prikažite naučeni model $h(\mathbf{x})=\mathbf{w}^\intercal\tilde{\mathbf{x}}$ i primjere za učenje. Izračunajte pogrešku učenja modela.

```python
def model(X, y, d):
    poly = PolynomialFeatures(d)
    X_tilda = poly.fit_transform(X)
    weights = np.matmul(linalg.pinv(X_tilda),y)
    return weights, X_tilda

def prediction_and_error(X_tilda, w, y):
    w = np.reshape(w,(-1,1))
    h = np.matmul(X_tilda,w)
    h = h.flatten()
    E = sklearn.metrics.mean_squared_error(y, h)
    return h,E
        
w, X_tilda = model(X, y, 3)
h, E = prediction_and_error(X_tilda, w, y)

plt.scatter(X, y)
plt.plot(X, h)
plt.show()
print("Greska je ", E)

```

```python
def fit_and_get_errors(X_train, X_test, y_train, y_test):
    
    Train_err=[]
    Test_err=[]
    for d in range(1, 21):
        w, X_tilda = model(X_train, y_train, d)
        h1, E_train = prediction_and_error(X_tilda, w, y_train)
        Train_err.append(np.log(E_train))

        poly = PolynomialFeatures(d)
        X_temp = poly.fit_transform(X_test)
        h2, E_test = prediction_and_error(X_temp, w, y_test)    
        Test_err.append(np.log(E_test))
    
    d_list = [i for i in range(1, 21)]
    return d_list, Train_err, Test_err

sigma = [100, 200, 500]
N = [50, 100, 500] #ovliko cemo uzmiat iz test/train skupa

fig, axs = plt.subplots(3, 4, figsize=(15, 15))
axs = axs.ravel()

X = make_instances(-5,5,1000)
X_train, X_test = train_test_split(X, test_size=0.5)

for i, std in enumerate(sigma):
    
    y = list(make_labels(X_train, f, std)) 
    y_train = np.array(y)
    
    y = list(make_labels(X_test,f,std)) 
    y_test = np.array(y)
    
    axs[i*4+3].scatter(X_train, y_train)
    axs[i*4+3].scatter(X_test, y_test)
    axs[i*4+3].set_title("Data for sigma= " + str(std))
    
    for j, n in enumerate(N):
        
        index_list = np.random.choice(np.arange(500), n, replace=False)
        X1_train = X_train[index_list]
        y1_train = y_train[index_list]
        
        index_list = np.random.choice(np.arange(500), n, replace=False)
        X1_test = X_test[index_list]
        y1_test = y_test[index_list]
        
        d,train,test=fit_and_get_errors(X1_train,X1_test,y1_train,y1_test)
        
        axs[i*4+j].set_title("N = " + str(n) + ", sigma = " + str(std))
        axs[i*4+j].plot(d,train,label="Train")
        axs[i*4+j].plot(d,test,label="Test")
        axs[i*4+j].legend()


plt.show()


```

Napravite izračun težina za regularizacijske faktore $\lambda=0$, $\lambda=1$ i $\lambda=10$ te usporedite dobivene težine.

**Q:** Kojih je dimenzija matrica koju treba invertirati?

```python
lambda_niz = [0, 1, 10]
for l in lambda_niz:
    
    X = np.array([[0],[1],[2],[4]])
    y = np.array([4,1,2,5])
    
    poly=PolynomialFeatures(3)
    psi=poly.fit_transform(X)
    
    temp = np.matmul(psi.transpose(), psi)
    temp = temp + l*np.identity(np.shape(temp)[0])
    temp = linalg.inv(temp)
    temp = np.matmul(temp, psi.transpose())
    w = np.matmul(temp, y)
    
    print(w)
```

### 6. L1-regularizacija i L2-regularizacija

Svrha regularizacije jest potiskivanje težina modela $\mathbf{w}$ prema nuli, kako bi model bio što jednostavniji. Složenost modela može se okarakterizirati normom pripadnog vektora težina $\mathbf{w}$, i to tipično L2-normom ili L1-normom. Za jednom trenirani model možemo izračunati i broj ne-nul značajki, ili L0-normu, pomoću sljedeće funkcije:

```python
def nonzeroes(coef, tol=1e-6): 
    return len(coef) - len(coef[np.isclose(0, coef, atol=tol)])
```

#### (a)

Za ovaj zadatak upotrijebite skup za učenje i skup za testiranje iz zadatka 3b. Trenirajte modele **L2-regularizirane** polinomijalne regresije stupnja $d=20$, mijenjajući hiperparametar $\lambda$ u rasponu $\{1,2,\dots,100\}$. Za svaki od treniranih modela izračunajte L{0,1,2}-norme vektora težina $\mathbf{w}$ te ih prikažite kao funkciju od $\lambda$.

**Q:** Objasnite oblik obiju krivulja. Hoće li krivulja za $\|\mathbf{w}\|_2$ doseći nulu? Zašto? Je li to problem? Zašto?

**Q:** Za $\lambda=100$, koliki je postotak težina modela jednak nuli, odnosno koliko je model rijedak?

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
import warnings

warnings.filterwarnings("ignore")

X = make_instances(-5, 5, 50)
y = list(make_labels(X, f, 200))
y = np.array(y)
poly = PolynomialFeatures(20)
psi = poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(psi, y, test_size=0.5)

l0 = []
l1 = []
l2 = []
for l in range(1, 101):
        clf = Ridge(alpha = l)
        clf.fit(X_train, y_train)
        #coefs = []
        #coefs.append(clf.intercept_)
        #coefs.append(clf.coef_)
        #coefs = coefs.flatten()
        #coefs = np.array(coefs)
        #print(coefs)
        
        coefs = clf.coef_
        coefs[0] = clf.intercept_ #nisan 100% sig jel ovo triba, al inace prvi koef bude uvik 0
        coefs = coefs.flatten()
        
        l0norm = nonzeroes(coefs)
        l1norm = np.sum(np.abs(coefs))
        l2norm = np.sqrt(np.matmul(coefs.transpose(), coefs))
        l0.append(l0norm)
        l1.append(l1norm)
        l2.append(l2norm)

plt.plot(range(1, 101), l0, label="L0")
plt.plot(range(1, 101), l1, label="L1")
plt.plot(range(1, 101), l2, label="L2")
plt.legend()
plt.show()
```

#### (b)

Glavna prednost L1-regularizirane regresije (ili *LASSO regression*) nad L2-regulariziranom regresijom jest u tome što L1-regularizirana regresija rezultira **rijetkim modelima** (engl. *sparse models*), odnosno modelima kod kojih su mnoge težine pritegnute na nulu. Pokažite da je to doista tako, ponovivši gornji eksperiment s **L1-regulariziranom** regresijom, implementiranom u klasi  [`Lasso`](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html) u modulu [`sklearn.linear_model`](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.linear_model).

```python
X = make_instances(-5, 5, 50)
y = list(make_labels(X, f, 200))
y = np.array(y)
poly = PolynomialFeatures(20)
psi = poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(psi, y, test_size=0.5)

l0 = []
l1 = []
l2 = []
for l in range(1, 101):
        clf = Lasso(alpha = l)
        clf.fit(X_train, y_train)
        #coefs = []
        #coefs.append(clf.intercept_)
        #coefs.extend(clf.coef_)
        #coefs = coefs.flatten()
        #coefs = np.array(coefs)
        #print(coefs)
        
        coefs = clf.coef_
        coefs[0] = clf.intercept_ 
        coefs = coefs.flatten()
        
        l0norm = nonzeroes(coefs)
        l1norm = np.sum(np.abs(coefs))
        l2norm = np.sqrt(np.matmul(coefs.transpose(), coefs))
        l0.append(l0norm)
        l1.append(l1norm)
        l2.append(l2norm)

plt.plot(range(1, 101), l0, label="L0")
plt.plot(range(1, 101), l1, label="L1")
plt.plot(range(1, 101), l2, label="L2")
plt.legend()
plt.show()
```

### 7. Značajke različitih skala

Često se u praksi možemo susreti sa podatcima u kojima sve značajke nisu jednakih magnituda. Primjer jednog takvog skupa je regresijski skup podataka `grades` u kojem se predviđa prosjek ocjena studenta na studiju (1--5) na temelju dvije značajke: bodova na prijamnom ispitu (1--3000) i prosjeka ocjena u srednjoj školi. Prosjek ocjena na studiju izračunat je kao težinska suma ove dvije značajke uz dodani šum.

Koristite sljedeći kôd kako biste generirali ovaj skup podataka.

```python
n_data_points = 500
np.random.seed(69)

# Generiraj podatke o bodovima na prijamnom ispitu koristeći normalnu razdiobu i ograniči ih na interval [1, 3000].
exam_score = np.random.normal(loc=1500.0, scale = 500.0, size = n_data_points) 
exam_score = np.round(exam_score)
exam_score[exam_score > 3000] = 3000
exam_score[exam_score < 0] = 0

# Generiraj podatke o ocjenama iz srednje škole koristeći normalnu razdiobu i ograniči ih na interval [1, 5].
grade_in_highschool = np.random.normal(loc=3, scale = 2.0, size = n_data_points)
grade_in_highschool[grade_in_highschool > 5] = 5
grade_in_highschool[grade_in_highschool < 1] = 1

# Matrica dizajna.
grades_X = np.array([exam_score,grade_in_highschool]).T

# Završno, generiraj izlazne vrijednosti.
rand_noise = np.random.normal(loc=0.0, scale = 0.5, size = n_data_points)
exam_influence = 0.9
grades_y = ((exam_score / 3000.0) * (exam_influence) + (grade_in_highschool / 5.0) \
            * (1.0 - exam_influence)) * 5.0 + rand_noise
grades_y[grades_y < 1] = 1
grades_y[grades_y > 5] = 5 
```

```python
plt.scatter(grades_X[:,0], grades_y)
plt.title("Ovisnost prosjeka ocjena na studiju o prijemnom ispitu")
plt.show()
plt.scatter(grades_X[:,1], grades_y)
plt.title("Ovisnost prosjeka ocjena na studiju o prosjeku u srednjoj skoli")
plt.show()
```
